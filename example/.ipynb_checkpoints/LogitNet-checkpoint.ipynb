```python
# loading required libraries
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glmnet.logistic import LogitNet
from sklearn.model_selection import train_test_split
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
```

```python
# loading example data from sklearn
data = load_breast_cancer()
X = data.data
y = data.target
cols = data.feature_names

# putting features (X) into Pandas DataFrame
X = pd.DataFrame(data=X, columns=cols)
X.head(2)
```

```python
# create stratified train/test splits
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    stratify=y,
                                                    random_state=1367)
```

```python
# train a LogitNet model
model = LogitNet(alpha=0.95, scoring="roc_auc", random_state=1367)
model.fit(X_train, y_train, )
```

```python
y_pred_proba = model.predict_proba(X_test)
y_pred_proba[:10]
```

```python
# plotting coeff. paths with log-yaxis
model.plot_coeff_path(yscale="log")
```

```python
# this can be done via plotting too!
from glmnet.plotting import coeff_path_plot
coeff_path_plot(model)
```

```python
# plot cross-validation scores
model.plot_cv_score()
```

```python
# this can be done via plotting too!
from glmnet.plotting import cv_score_plot
cv_score_plot(model)
```
